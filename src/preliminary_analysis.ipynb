```python
import json

import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
file = open('../data/train-claims.json')
train_data = json.load(file)
file.close()

file = open('../data/evidence.json')
evidence = json.load(file)
file.close()

file = open('../data/dev-claims.json')
dev_data = json.load(file)
file.close()
```

```python
print("Number of training instances: ", len(train_data))
print("Number of dev set instances: ", len(dev_data))
print("Number of evidences: ", len(evidence))
```

```python
evidence_dist = []
labels = []
claim_length = []

for claim_no, value in train_data.items():
    claim_length.append(len(value['claim_text']))
    evidence_dist.append(len(value['evidences']))
    labels.append(value['claim_label'])

print("Most # of evidence: ", max(evidence_dist))
print("Least # of evidence: ", min(evidence_dist))
print("Average number of evidence per claim: ", sum(evidence_dist)/len(evidence_dist))
```

```python
evidence_dist = Counter(evidence_dist)
plt.bar(evidence_dist.keys(), evidence_dist.values())
plt.title("Distribution of number of evidences")
plt.show()
```

```python
label_dist = Counter(labels)
plt.bar(label_dist.keys(), label_dist.values())
plt.title('Distribution of labels')
plt.show()
```
